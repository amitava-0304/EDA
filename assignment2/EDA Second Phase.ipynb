```python
## Data Analysis Phase
## MAin aim is to understand more about the data

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
## Display all the columns of the dataframe

pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('dataset/googleplaystore.csv')
```

```python
#Find null values

dataset.isnull().sum()
```

```python
def printinfo():
    temp = pd.DataFrame(index=dataset.columns)
    temp['data_type'] = dataset.dtypes
    temp['null_count'] = dataset.isnull().sum()
    temp['unique_count'] = dataset.nunique()
    return temp
```

# Since there is only one missing value in this column, So, letâ€™s fill the missing value. After cross-checking in the play store the missing value is found to be Free, So now we can fill the missing value with free 

```python
dataset['Type'].fillna("Free", inplace = True)
```

```python
dataset.isnull().sum()
```

# Now, the reviews column has been converted to int type, so now we can move to the Column: Size Converting the Size Column from object to integer, but this column contains some of the special characters like  M , K & also it has a some of the value as Varies with device . We need to remove all of these and then convert it to int or float .

```python
dataset['Size'] = dataset.Size.apply(lambda x: x.replace('M', 'e+6'))# For converting the M to Mega
```

```python
dataset.Size.unique()
```

```python
dataset['Size'] = dataset.Size.apply(lambda x: x.replace('k', 'e+3'))# For convertinf the K to Kilo
```

```python
dataset.Size.unique()
```

```python
dataset['Size'] = dataset.Size.replace('Varies with device', np.NaN)
```

```python
dataset['Size'] = pd.to_numeric(dataset['Size']) # Converting the string to Numeric type
```

```python
printinfo()
```

Since we converted the Varies with device value to Nan , so we have to do something with those set of Nan values data. It will be a better idea to drop the Rows of the column Size having Nanvalues because it will be not an efficient idea to replace those values with mean or mode since the size of some apps would be too large and some of them too small.

```python
dataset.dropna(subset = ['Size'], inplace=True)
```

```python
Column: Installs :
To convert this column from object to integer type. First of all, we will need to remove the +symbol from these values.
```

```python
dataset['Installs'] = dataset['Installs'].map(lambda x: x.rstrip('+'))
dataset['Installs'] = dataset['Installs'].replace(',', '', regex = True)
dataset['Installs'] = dataset['Installs'].replace('Free', '', regex = True)
dataset['Installs'] = dataset['Installs'].replace('', 0, regex = True)
dataset['Installs'] = dataset['Installs'].astype(int)
```

```python
printinfo()
```

```python
So, now we are only left with the Price column.
Column: Price :
Converting this column from objectto Numeric type.
```

```python
dataset['Price'].value_counts()
```

```python
The values contain a special symbol $ which can be removed and then converted to the numeric type.
```

```python
dataset['Price'] =dataset.Price.apply(lambda x: x.strip('$'))
```

```python
dataset['Price'] = pd.to_numeric(dataset['Price'])
```

```python
printinfo()
```

```python
# Start of cleaning
# There were given some commands to locate any invalid data
# I noticed that are some misplacing, e.g. here, "M" should move from "Reviews"
```

```python
dataset['Reviews'] = dataset['Reviews'].map(lambda x: x.rstrip('M'))
```

```python
dataset.Reviews= dataset.Reviews.astype(float) 
```

```python
printinfo()
```
